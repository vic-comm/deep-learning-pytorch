```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Restaurant_revenue (1).csv')
```

```python
data
```

```python
data.describe()
```

```python
data['Cuisine_Type'].value_counts()
```

```python
data.isnull().sum()
```

```python
data.groupby('Cuisine_Type').agg({'Number_of_Customers':[np.mean, np.sum, np.std], 'Menu_Price':[np.mean, np.sum, np.std, np.var], 'Monthly_Revenue':[np.mean, np.sum, np.std, np.var]})
```

```python
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# map = {'Japanese':1, 'American':2, 'Mexican':3, 'Italian':4}
# data['Cuisine_Type'] = data['Cuisine_Type'].map(map)
```

```python
data.info()
```

```python
# One hot encoding
data = pd.get_dummies(data, columns=['Cuisine_Type'])
```

```python
data
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x = data.drop('Monthly_Revenue', axis=1)
y = data['Monthly_Revenue']
```

```python
scaler = StandardScaler()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=17, test_size=0.3)
```

```python
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
reg = Ridge()
neighbor = KNeighborsRegressor()
```

```python
reg.fit(x_train, y_train)
```

```python
neighbor.fit(x_train, y_train)
```

```python
prediction = reg.predict(x_train)
mae = metrics.mean_absolute_error(prediction, y_train)
mape = metrics.mean_absolute_percentage_error(prediction, y_train)
r2_score = metrics.r2_score(prediction, y_train)
print('Ridge training score on raw data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
prediction = neighbor.predict(x_train)
mae = metrics.mean_absolute_error(prediction, y_train)
mape = metrics.mean_absolute_percentage_error(prediction, y_train)
r2_score = metrics.r2_score(prediction, y_train)
print('KNeighbots training score on raw data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
prediction = reg.predict(x_test)
mae = metrics.mean_absolute_error(prediction, y_test)
mape = metrics.mean_absolute_percentage_error(prediction, y_test)
r2_score = metrics.r2_score(prediction, y_test)
print('Ridge testing score on raw data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
prediction = neighbor.predict(x_test)
mae = metrics.mean_absolute_error(prediction, y_test)
mape = metrics.mean_absolute_percentage_error(prediction, y_test)
r2_score = metrics.r2_score(prediction, y_test)
print('kNeighbor testing score on raw data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
reg2 = reg.fit(x_train_scaled, y_train)
```

```python
neighbor2 = neighbor.fit(x_train_scaled, y_train)
```

```python
prediction = neighbor2.predict(x_train_scaled)
mae = metrics.mean_absolute_error(prediction, y_train)
mape = metrics.mean_absolute_percentage_error(prediction, y_train)
r2_score = metrics.r2_score(prediction, y_train)
print('KNeighbots training score on scaled data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
prediction = reg2.predict(x_test_scaled)
mae = metrics.mean_absolute_error(prediction, y_test)
mape = metrics.mean_absolute_percentage_error(prediction, y_test)
r2_score = metrics.r2_score(prediction, y_test)
print('Ridge testing score on scaled data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```

```python
prediction = neighbor2.predict(x_test_scaled)
mae = metrics.mean_absolute_error(prediction, y_test)
mape = metrics.mean_absolute_percentage_error(prediction, y_test)
r2_score = metrics.r2_score(prediction, y_test)
print('kNeighbor testing score on scaled data')
print(f"mean_absolute_error: {mae}")
print(f"mean_absolute_percentage_error: {mape}")
print(f"R squared: {r2_score}")
```
